```python
dfQ1=pd.concat([df_hagush_hagadol, df_florentin, df_neve_ofer, df_north_star, df_kiryat_shalom, df_neve_shaanan], axis=0)

#dfQ1
```

```python
dfQ1['fam']=0
for i in range (5712):
    if(dfQ1.iloc[i, 13] <3 or dfQ1.iloc[i, 12]<3 or dfQ1.iloc[i, 11]<2.5 or dfQ1.iloc[i, 10]<2.45):
        dfQ1.iloc[i, 14] = 0
    else:
        dfQ1.iloc[i, 14] = 1
#dfQ1
```

```python
X = dfQ1.drop(['fam'], axis=1)
y = dfQ1['fam']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
clf = LogisticRegression(solver='lbfgs', max_iter=20000) # max_iter is the number of itrations it takes for all centroids to converge
clf.fit(xtrain, ytrain)

y_pred = clf.predict(xtest)
```

```python
print("Accuracy is:",metrics.accuracy_score(ytest, y_pred))
```
